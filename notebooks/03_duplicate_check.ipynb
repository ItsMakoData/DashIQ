```python
import pandas as pd
from pathlib import Path
```

```python
# Get the current notebook directory
notebook_dir = Path.cwd()  # This points to the folder where the notebook is running

# Construct the correct path to your CSV
data_path = notebook_dir / "../data/raw/sales_records.csv"

# Resolve full absolute path (optional, just for debugging)
data_path = data_path.resolve()
print("Reading from:", data_path)

# Load dataset
df = pd.read_csv(data_path)

# Show first 5 rows
df.head()
```

```python
# Count exact duplicate rows
duplicate_count = df.duplicated().sum()
duplicate_count
```

```python
# Show duplicate rows
duplicate_rows = df[df.duplicated()]
duplicate_rows.head()
```

```python
# Count duplicate Order IDs
duplicate_order_ids = df["Order ID"].duplicated().sum()
duplicate_order_ids
```

```python
# Show rows with repeated Order IDs (For Removal)
repeated_orders = df[df["Order ID"].duplicated(keep=False)].sort_values("Order ID")
repeated_orders.head(10)
```

```python
# Create a flag for duplicate
# Mark exact duplicate rows
df["is_duplicate"] = df.duplicated()

# Count flagged duplicates
df["is_duplicate"].value_counts()
```

```python
# Marked for removal
rows_marked_for_removal = df[df["is_duplicate"] == True]
rows_marked_for_removal.head()
```

```python
# Saving duplicate rows for audit trail
output_path = data_path.parent / "duplicate_rows_review.csv"
rows_marked_for_removal.to_csv(output_path, index=False)
```
