```python
import pandas as pd
import sqlite3
```

```python
# Load CSV
df = pd.read_csv("../data/cleaned/sales_records_enhanced.csv")

# Connect DB
conn = sqlite3.connect("../data/cleaned/dashiq.db")
```

```python
python_revenue = df["Total Revenue"].sum()
python_revenue
```

```python
sql_revenue = pd.read_sql("""
SELECT SUM("Total Revenue") AS total_revenue
FROM fact_sales
""", conn)

sql_revenue
```

```python
print("Python:", python_revenue)
print("SQL:", sql_revenue["total_revenue"][0])
```

```python
python_profit = df["Total Profit"].sum()
```

```python
sql_profit = pd.read_sql("""
SELECT SUM("Total Profit") AS total_profit
FROM fact_sales
""", conn)
```

```python
print("Python:", python_profit)
print("SQL:", sql_profit["total_profit"][0])
```

```python
df_2020 = df[df["Year"] == 2013]
df_2020["Total Revenue"].sum()
```

```python
pd.read_sql("""
SELECT SUM("Total Revenue") AS total_revenue
FROM fact_sales
WHERE Year = 2013
""", conn)
```

```python
python_margin = df["Total Profit"].sum() / df["Total Revenue"].sum()
python_margin
```

```python
sql_margin = pd.read_sql("""
SELECT 
    SUM("Total Profit") * 1.0 / SUM("Total Revenue") AS profit_margin
FROM fact_sales
""", conn)

sql_margin
```
