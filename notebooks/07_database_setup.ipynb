```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("../data/cleaned/sales_records_enhanced.csv")

df.head()
```

```python
conn = sqlite3.connect("../data/cleaned/dashiq.db")

print("Database created successfully")
```

```python
df.to_sql("fact_sales", conn, if_exists="replace", index=False)

print("Table 'fact_sales' created")
```

```python
# dim_region
df[["Region"]].drop_duplicates().to_sql("dim_region", conn, if_exists="replace", index=False)

# dim_country
df[["Country"]].drop_duplicates().to_sql("dim_country", conn, if_exists="replace", index=False)

# dim_product
df[["Item Type"]].drop_duplicates().to_sql("dim_product", conn, if_exists="replace", index=False)

# dim_channel
df[["Sales Channel"]].drop_duplicates().to_sql("dim_channel", conn, if_exists="replace", index=False)

# dim_priority
df[["Order Priority"]].drop_duplicates().to_sql("dim_priority", conn, if_exists="replace", index=False)

# dim_date
df[["Order Date", "Year", "Month"]].drop_duplicates().to_sql("dim_date", conn, if_exists="replace", index=False)

print("Dimension tables created")
```
