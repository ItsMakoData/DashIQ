```python
import pandas as pd
```

```python
df = pd.read_csv("../data/cleaned/sales_records_enhanced.csv")

df["Order Date"] = pd.to_datetime(df["Order Date"])
df.head()
```

```python
monthly_sales = df.groupby(["Year", "Month"])["Total Revenue"].sum().reset_index()
monthly_sales.head()
```

```python
category_perf = df.groupby("Item Type")["Total Revenue"].sum().sort_values(ascending=False)
category_perf

#Total Profit by Item Type
df.groupby("Item Type")["Total Profit"].sum().sort_values(ascending=False)
#Units Sold by Item Type
df.groupby("Item Type")["Units Sold"].sum().sort_values(ascending=False)
#Revenue by Region
region_perf = df.groupby("Region")["Total Revenue"].sum().sort_values(ascending=False)
region_perf
#Profit by Region
df.groupby("Region")["Total Profit"].sum().sort_values(ascending=False)
#Revenue by discount bucket
df.groupby("Discount Bucket")["Total Revenue"].sum()
df.groupby("Discount Bucket")["Total Profit"].sum()
df.groupby("Discount Bucket")["Profit Margin"].mean()
```
